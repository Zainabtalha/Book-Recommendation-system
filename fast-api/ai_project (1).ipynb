```python
import numpy as np
import pandas as pd
```

```python
ratings = pd.read_csv('Ratings.csv')
users = pd.read_csv('Users.csv')
books = pd.read_csv("Books.csv")
```

```python
# merging two different dataframes, books and ratings, on isbn columnmb
```

```python
mergedTable = ratings.merge(books,on='ISBN')
mergedTable.head()
```

```python
Rating_Num = mergedTable.groupby('Book-Title').count()['Book-Rating'].reset_index()
Rating_Num.rename(columns = {'Book-Rating':'Rating Count'},inplace=True)
```

```python
Rating_Num
```

```python
Avg_Rating_Num=mergedTable.groupby('Book-Title')['Book-Rating'].mean().reset_index()
Avg_Rating_Num.rename(columns = {'Book-Rating':'Avg_Rating'},inplace=True)
```

```python
popularity = Avg_Rating_Num.merge(Rating_Num,on='Book-Title')
popularity
```

```python
popularity=popularity[popularity['Rating Count']>=250].sort_values('Avg_Rating',ascending=False).head(50)
```

```python
popularity = popularity.merge(books,on='Book-Title').drop_duplicates('Book-Title')
```

```python
popularity=popularity[['Book-Title','Book-Author','Image-URL-M','Rating Count','Avg_Rating']]
popularity
```

```python
popularity.drop_duplicates('Book-Title',inplace=True)
```

```python
#Starting collaborative filtering approach
```

```python
#have to represent the data in a matrix form where books are rows and users are columns and their intersection is rating
```

```python
mergedTable
```

```python
temp = mergedTable.groupby("User-ID").count()['Book-Rating']>200
temp = temp[temp]
# users who gave rating to atleast 200 books
```

```python
finalUsers = temp.index
finalUsers#users who gave rating to atleast 200 books
```

```python
filtered_rating=mergedTable[mergedTable['User-ID'].isin(finalUsers)]
```

```python
y=filtered_rating.groupby("Book-Title").count()['Book-Rating']>50 # books on which there are more than 50 ratings
famous_books=y[y].index
```

```python
final=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pivot_table=final.pivot_table(index="Book-Title",columns="User-ID",values="Book-Rating")
```

```python
pivot_table.fillna(0,inplace=True)
```

```python
pivot_table.index
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
score=cosine_similarity(pivot_table)
```

```python
def recommend(book_name):
  index=np.where(pivot_table.index==book_name)[0][0]
  distances=sorted(list(enumerate(score[index])),key=lambda x:x[1],reverse=True)[1:6]

  data = {}
  j=0
  for i in distances:
    item = {}
    temp_df = books[books['Book-Title']==pivot_table.index[i[0]]]
    item['Book Title']=(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values)[0])
    item['Book Author']=(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values)[0])
    item['Book Image']=(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values)[0])

    data[j]=item
    j=j+1
  return data
```

```python
with open('pivotTable.pkl', 'wb') as file:
    pickle.dump(pivot_table, file)

with open('score.pkl', 'wb') as file:
    pickle.dump(score, file)

with open('books.pkl', 'wb') as file:
    pickle.dump(books, file)

```
